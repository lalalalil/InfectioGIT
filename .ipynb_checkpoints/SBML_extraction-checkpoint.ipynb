```python
# récuperer tous les fichiers et pas que les "modèles" avec commentaire

'''
on trouve 1471 fichiers pour 144 modèles

repartition :
- 53 modèles covid
- 27 modèles HIV
- 27 modèles influenza
- 27 modèles tuverculosis
- 1 modèle m-pox
'''

import shutil
import json
import time  # Essential for rate-limiting to avoid HTTP 403 Forbidden errors
from pathlib import Path
from bioservices import BioModels
import biomodels

# Root directory configuration
racine_data = Path("./BioModels_Data") 
racine_data.mkdir(parents=True, exist_ok=True)

s = BioModels()

# Dictionary mapping local folder names to search keywords (tuples)
# Using multiple keywords increases the scope of the search per category
diseases = {
    "dengue_files": ('dengue','DENV'),
    "chikungunya_files": ("chikungunya", "CHIKV"),
    "lyme_files": ('lyme', 'borrelia', 'borreliosis'),
    "mpox_files": ('mpox', 'monkeypox'),
    "west_nile_files": ('west nile', 'WNV'),
    "influenza_files": ("influenza", "influenza virus","avian influenza", "H5N1"),
    "tuberculosis_files": ("tuberculosis", "TB", "mycobacterium"),
    "hiv_files": ("HIV", "human immunodeficiency virus"),
    "covid_files": ("covid", "SARS-CoV-2")
}

# Iterate through the dictionary items to keep queries grouped by folder name
for folder_name, queries in diseases.items():
    print(f"\n--- Searching for: {folder_name} ---")
    
    # Combine keywords into a single Boolean OR query for efficiency
    query_string = " OR ".join(f'"{q}"' for q in queries)
    results = s.search(query_string, numResults=100) 
    
    # Security check: If the server returns an error object instead of a dict, 
    # we catch it here to prevent 'AttributeError' and wait before retrying.
    if not isinstance(results, dict):
        print(f"Server error for {folder_name} (Possible 403 Forbidden). Waiting...")
        time.sleep(5)
        continue

    # Extract model IDs from the results list
    ids = [res['id'] for res in results.get('models', [])]
    time.sleep(1) # Safety delay between category searches

    for model_id in ids:
        try:
            # 1. Create a subdirectory specific to the Model ID inside the disease folder
            dossier_model = racine_data / folder_name / model_id
            dossier_model.mkdir(parents=True, exist_ok=True)

            # 2. Retrieve metadata (a list of all available file objects for this model)
            files_objects = biomodels.get_metadata(model_id)
            if not files_objects:
                continue
            
            # Save the JSON metadata locally for future reference/parsing
            meta_path = dossier_model / f"metadata.json"
            with open(meta_path, "w", encoding="utf-8") as f_meta:
                json.dump(files_objects, f_meta, indent=4, default=str)

            print(f"Downloading files for {model_id}...")
            
            # 3. Comprehensive retrieval: Loop through EVERY file listed in the metadata
            # This ensures we get .xml, .m, .cps, .owl, .pdf, etc.
            for target_file_obj in files_objects:
                try:
                    # Extract the real filename from the object attributes
                    nom_reel = getattr(target_file_obj, 'name', str(target_file_obj))
                    
                    if not nom_reel or nom_reel == "None":
                        continue

                    # Download the file content/path
                    result = biomodels.get_file(target_file_obj)
                    dest_path = dossier_model / nom_reel

                    # If the result is a local temp path, move/copy it
                    if isinstance(result, (str, Path)) and Path(result).exists():
                        shutil.copy(result, dest_path)
                    else:
                        # If result is raw data, write it directly
                        with open(dest_path, "w", encoding="utf-8") as f:
                            f.write(str(result))
                    
                    print(f"  + {nom_reel} retrieved")
                except Exception as e_file:
                    # Individual file failure shouldn't stop the whole model download
                    print(f"  ! Error on file {nom_reel}: {e_file}")
            
            # Crucial Sleep: Throttles requests to avoid IP bans during bulk downloads
            time.sleep(0.8)

        except Exception as e:
            print(f"Error on model {model_id}: {e}")
```

```python
'''
téléchargement du JSON de la page web et le tri automatique dans les sous-répertoires metadata et model.
'''

import shutil
import json
import time
import requests  # Nécessaire pour le format ?format=json
from pathlib import Path
from bioservices import BioModels
import biomodels

# Configuration du dossier racine
racine_data = Path("./BioModels_Data_json") 
racine_data.mkdir(parents=True, exist_ok=True)

s = BioModels()
diseases = [
    'dengue', 'DENV', 
    'chikungunya', 'CHIKV', 
    'lyme', 'borrelia', 'borreliosis', 
    'mpox', 'monkeypox', 
    'west nile', 'WNV', 
    'influenza', 'influenza virus', 'avian influenza', 'H5N1', 
    'tuberculosis', 'TB', 'mycobacterium', 
    'HIV', 'human immunodeficiency virus', 
    'covid', 'SARS-CoV-2'
]

for disease in diseases:
    print(f"\n--- Recherche pour : {disease} ---")
    results = s.search(disease, numResults=100) 
    
    if not isinstance(results, dict):
        print(f"Erreur serveur pour {disease}. On patiente...")
        time.sleep(5)
        continue

    ids = [res['id'] for res in results.get('models', [])]
    time.sleep(1)

    for model_id in ids:
        try:
            # 1. Création de l'arborescence demandée
            dossier_base = racine_data / disease / model_id
            d_metadata = dossier_base / "metadata"
            d_model = dossier_base / "model"
            
            d_metadata.mkdir(parents=True, exist_ok=True)
            d_model.mkdir(parents=True, exist_ok=True)

            # 2. Téléchargement du JSON de la page web (format spécifié)
            print(f"Récupération JSON Web pour {model_id}...")
            web_url = f"https://www.ebi.ac.uk/biomodels/{model_id}?format=json"
            try:
                web_res = requests.get(web_url, timeout=10)
                if web_res.status_code == 200:
                    with open(d_metadata / f"{model_id}_web_page.json", "w", encoding="utf-8") as f_web:
                        json.dump(web_res.json(), f_web, indent=4)
            except:
                print(f"  ! Échec JSON Web")

            # 3. Récupération des métadonnées de fichiers
            files_objects = biomodels.get_metadata(model_id)
            if not files_objects:
                continue
            
            # Sauvegarde dans le dossier metadata
            with open(d_metadata / "files_list.json", "w", encoding="utf-8") as f_meta:
                json.dump(files_objects, f_meta, indent=4, default=str)

            # 4. Récupération et tri des fichiers
            for target_file_obj in files_objects:
                try:
                    nom_reel = getattr(target_file_obj, 'name', str(target_file_obj))
                    if not nom_reel or nom_reel == "None": continue

                    # Logique de tri
                    if "metadata" in nom_reel.lower() or nom_reel.endswith('.json') or nom_reel.endswith('.rdf'):
                        dest_path = d_metadata / nom_reel
                    else:
                        dest_path = d_model / nom_reel

                    result = biomodels.get_file(target_file_obj)

                    if isinstance(result, (str, Path)) and Path(result).exists():
                        shutil.copy(result, dest_path)
                    else:
                        with open(dest_path, "w", encoding="utf-8") as f:
                            f.write(str(result))
                    
                    print(f"  + {nom_reel} -> {dest_path.parent.name}")
                except Exception as e_file:
                    print(f"  ! Erreur fichier {nom_reel}")
            
            time.sleep(0.8)

        except Exception as e:
            print(f"Erreur sur {model_id}: {e}")
```

```python
'''
import csv, un dictionnaire stats_extensions pour compter les fichiers, et une étape de nettoyage 
pour supprimer les fichiers du dossier metadata qui ne sont pas les deux fichiers JSON générés par le script.
'''

import shutil
import json
import time
import requests
import csv  # Added for CSV generation
from pathlib import Path
from bioservices import BioModels
import biomodels

# Root directory configuration
racine_data = Path("./BioModels_Data_cleaned_json_dico") 
racine_data.mkdir(parents=True, exist_ok=True)

s = BioModels()

diseases = {
    "dengue_files": ('dengue','DENV'),
    "chikungunya_files": ("chikungunya", "CHIKV"),
    "lyme_files": ('lyme', 'borrelia', 'borreliosis'),
    "mpox_files": ('mpox', 'monkeypox'),
    "west_nile_files": ('west nile', 'WNV'),
    "influenza_files": ("influenza", "influenza virus","avian influenza", "H5N1"),
    "tuberculosis_files": ("tuberculosis", "TB", "mycobacterium"),
    "hiv_files": ("HIV", "human immunodeficiency virus"),
    "covid_files": ("covid", "SARS-CoV-2")
}

# List to store data for the final CSV
csv_data = []

for folder_name, queries in diseases.items():
    print(f"\n--- Category: {folder_name} ---")
    query_string = " OR ".join(f'"{q}"' for q in queries)
    results = s.search(query_string, numResults=100) 
    
    if not isinstance(results, dict):
        print(f"Server error for {folder_name}. Waiting...")
        time.sleep(5)
        continue

    ids = [res['id'] for res in results.get('models', [])]
    time.sleep(1)

    for model_id in ids:
        try:
            dossier_base = racine_data / folder_name / model_id
            d_metadata = dossier_base / "metadata"
            d_model = dossier_base / "model"
            
            d_metadata.mkdir(parents=True, exist_ok=True)
            d_model.mkdir(parents=True, exist_ok=True)

            # Keep track of generated filenames for cleaning
            generated_json = [f"{model_id}_web_metadata.json", "files_list.json"]

            # 3. Download Web JSON
            print(f"  > Fetching Web JSON for {model_id}...")
            web_url = f"https://www.ebi.ac.uk/biomodels/{model_id}?format=json"
            try:
                web_res = requests.get(web_url, timeout=10)
                if web_res.status_code == 200:
                    with open(d_metadata / generated_json[0], "w", encoding="utf-8") as f_web:
                        json.dump(web_res.json(), f_web, indent=4)
            except Exception:
                print(f"    ! Web JSON failed")

            # 4. Retrieve files list
            files_objects = biomodels.get_metadata(model_id)
            if not files_objects: continue
            
            with open(d_metadata / generated_json[1], "w", encoding="utf-8") as f_meta:
                json.dump(files_objects, f_meta, indent=4, default=str)

            # Dictionary to count extensions for THIS model
            stats_extensions = {}

            # 5. Download and sort files
            print(f"  > Downloading files for {model_id}...")
            for target_file_obj in files_objects:
                try:
                    nom_reel = getattr(target_file_obj, 'name', str(target_file_obj))
                    if not nom_reel or nom_reel == "None": continue

                    # Count extension
                    ext = Path(nom_reel).suffix.lower() or "no_extension"
                    stats_extensions[ext] = stats_extensions.get(ext, 0) + 1

                    if any(x in nom_reel.lower() for x in ["metadata", ".json", ".rdf", ".owl"]):
                        dest_path = d_metadata / nom_reel
                    else:
                        dest_path = d_model / nom_reel

                    result = biomodels.get_file(target_file_obj)
                    if isinstance(result, (str, Path)) and Path(result).exists():
                        shutil.copy(result, dest_path)
                    else:
                        with open(dest_path, "w", encoding="utf-8") as f:
                            f.write(str(result))
                except Exception:
                    print(f"    ! Error on file {nom_reel}")

            # 6. Cleaning: Remove metadata files that were NOT generated by the code
            for file_path in d_metadata.iterdir():
                if file_path.name not in generated_json:
                    file_path.unlink()

            # Prepare CSV row
            row = {"category": folder_name, "model_id": model_id}
            row.update(stats_extensions)
            csv_data.append(row)

            time.sleep(0.8)

        except Exception as e:
            print(f"Error processing {model_id}: {e}")

# 7. Write the CSV file
if csv_data:
    # Get all unique extensions found to create CSV headers
    all_extensions = sorted(list(set().union(*(d.keys() for d in csv_data if isinstance(d, dict)))))
    with open(racine_data / "summary_extensions.csv", "w", newline="", encoding="utf-8") as f_csv:
        writer = csv.DictWriter(f_csv, fieldnames=all_extensions)
        writer.writeheader()
        writer.writerows(csv_data)
    print(f"\nCSV Summary generated: {racine_data}/summary_extensions.csv")
```

```python
'''
liste d'extensions considérées comme "non-modélisation" (images, documents, notes)
qui seront redirigées vers le dossier metadata
'''

import shutil
import json
import time
import requests
import csv
from pathlib import Path
from bioservices import BioModels
import biomodels

# Root directory configuration
racine_data = Path("./BioModels_Data_cleaned_redirection_metadata_json_dico_") 
racine_data.mkdir(parents=True, exist_ok=True)

s = BioModels()

diseases = {
    "dengue_files": ('dengue','DENV'),
    "chikungunya_files": ("chikungunya", "CHIKV"),
    "lyme_files": ('lyme', 'borrelia', 'borreliosis'),
    "mpox_files": ('mpox', 'monkeypox'),
    "west_nile_files": ('west nile', 'WNV'),
    "influenza_files": ("influenza", "influenza virus","avian influenza", "H5N1"),
    "tuberculosis_files": ("tuberculosis", "TB", "mycobacterium"),
    "hiv_files": ("HIV", "human immunodeficiency virus"),
    "covid_files": ("covid", "SARS-CoV-2")
}

# Define extensions that should be moved to 'metadata' instead of 'model'
EXTENSIONS_METADATA = {'.png', '.jpg', '.jpeg', '.pdf', '.txt', '.docx', '.doc', '.xlsx', '.xls', '.csv'}

# List to store data for the final CSV report
csv_data = []

for folder_name, queries in diseases.items():
    print(f"\n--- Category: {folder_name} ---")
    query_string = " OR ".join(f'"{q}"' for q in queries)
    results = s.search(query_string, numResults=100) 
    
    if not isinstance(results, dict):
        print(f"Server error for {folder_name}. Waiting...")
        time.sleep(5)
        continue

    ids = [res['id'] for res in results.get('models', [])]
    time.sleep(1)

    for model_id in ids:
        try:
            # Create local directory structure
            dossier_base = racine_data / folder_name / model_id
            d_metadata = dossier_base / "metadata"
            d_model = dossier_base / "model"
            
            d_metadata.mkdir(parents=True, exist_ok=True)
            d_model.mkdir(parents=True, exist_ok=True)

            # Whitelist of files to keep in the metadata folder (prevent deletion during cleanup)
            files_to_preserve = [f"{model_id}_web_metadata.json", "files_list.json"]

            # 3. Download Web JSON Metadata
            print(f"  > Fetching Web JSON for {model_id}...")
            web_url = f"https://www.ebi.ac.uk/biomodels/{model_id}?format=json"
            try:
                web_res = requests.get(web_url, timeout=10)
                if web_res.status_code == 200:
                    with open(d_metadata / files_to_preserve[0], "w", encoding="utf-8") as f_web:
                        json.dump(web_res.json(), f_web, indent=4)
            except Exception:
                print(f"    ! Web JSON fetch failed")

            # 4. Retrieve internal files list from BioModels
            files_objects = biomodels.get_metadata(model_id)
            if not files_objects: continue
            
            with open(d_metadata / files_to_preserve[1], "w", encoding="utf-8") as f_meta:
                json.dump(files_objects, f_meta, indent=4, default=str)

            # Dictionary to count extensions for the current model
            stats_extensions = {}

            # 5. Download and sort files based on extension and name
            print(f"  > Downloading and sorting files for {model_id}...")
            for target_file_obj in files_objects:
                try:
                    nom_reel = getattr(target_file_obj, 'name', str(target_file_obj))
                    if not nom_reel or nom_reel == "None": continue

                    # Identify file extension
                    ext = Path(nom_reel).suffix.lower() or "no_extension"
                    stats_extensions[ext] = stats_extensions.get(ext, 0) + 1

                    # Sorting logic:
                    # Move to metadata if: specific keywords in name OR extension in EXTENSIONS_METADATA
                    is_metadata_keyword = any(x in nom_reel.lower() for x in ["metadata", ".json", ".rdf", ".owl"])
                    is_non_modeling_ext = ext in EXTENSIONS_METADATA
                    
                    if is_metadata_keyword or is_non_modeling_ext:
                        dest_path = d_metadata / nom_reel
                        files_to_preserve.append(nom_reel) # Add to whitelist to avoid deletion
                    else:
                        dest_path = d_model / nom_reel

                    # Download the file
                    result = biomodels.get_file(target_file_obj)
                    if isinstance(result, (str, Path)) and Path(result).exists():
                        shutil.copy(result, dest_path)
                    else:
                        with open(dest_path, "w", encoding="utf-8") as f:
                            f.write(str(result))
                except Exception:
                    print(f"    ! Error processing file: {nom_reel}")

            # 6. Final Cleanup: Remove any unexpected files from the metadata folder
            for file_path in d_metadata.iterdir():
                if file_path.name not in files_to_preserve:
                    file_path.unlink()

            # Prepare statistics row for the final CSV
            row = {"category": folder_name, "model_id": model_id}
            row.update(stats_extensions)
            csv_data.append(row)

            time.sleep(0.8) # Anti-throttling pause

        except Exception as e:
            print(f"Critical error on model {model_id}: {e}")

# 7. Generate the global CSV summary
if csv_data:
    # Identify all unique extensions found across all models for CSV headers
    all_cols = sorted(list(set().union(*(d.keys() for d in csv_data if isinstance(d, dict)))))
    with open(racine_data / "summary_extensions.csv", "w", newline="", encoding="utf-8") as f_csv:
        writer = csv.DictWriter(f_csv, fieldnames=all_cols)
        writer.writeheader()
        writer.writerows(csv_data)
    print(f"\nCSV Summary generated successfully at: {racine_data}/summary_extensions.csv")
```

```python
'''modelling approach no modif struct'''

import json
from pathlib import Path

# Configuration du chemin racine
# Adaptez le nom du dossier à votre structure réelle
root_path = Path("./BioModels_Data_cleaned_json_dico")

def extract_modelling_approaches(base_dir):
    print(f"{'DISEASE':<20} | {'MODEL ID':<15} | {'MODELLING APPROACH'}")
    print("-" * 70)

    # On parcourt les dossiers de catégories (ex: covid_files)
    for category_dir in base_dir.iterdir():
        if category_dir.is_dir():
            
            # On parcourt les dossiers de modèles (ex: BIOMD...)
            for model_dir in category_dir.iterdir():
                if model_dir.is_dir():
                    
                    # Chemin vers le dossier metadata
                    metadata_dir = model_dir / "metadata"
                    model_id = model_dir.name
                    disease = category_dir.name
                    
                    if metadata_dir.exists():
                        # On cherche le fichier .json qui contient les métadonnées web
                        # Souvent nommé [ID]_web_metadata.json d'après vos scripts précédents
                        json_files = list(metadata_dir.glob("*_web_metadata.json"))
                        
                        for json_file in json_files:
                            try:
                                with open(json_file, 'r', encoding='utf-8') as f:
                                    data = json.load(f)
                                    
                                    # Extraction de la section modellingApproach
                                    # On utilise .get() pour éviter les erreurs si la clé est absente
                                    approach_data = data.get("modellingApproach", {})
                                    approach_name = approach_data.get("name", "Not specified")
                                    
                                    print(f"{disease:<20} | {model_id:<15} | {approach_name}")
                                    
                            except Exception as e:
                                print(f"Error reading {json_file.name}: {e}")

if __name__ == "__main__":
    if root_path.exists():
        extract_modelling_approaches(root_path)
    else:
        print(f"Directory {root_path} not found.")
```

```python
#output .csv modelling approach

import json
import csv
from pathlib import Path

# Configuration du chemin racine
root_path = Path("./BioModels_Data_cleaned_json_dico")
output_file = Path("modelling_approaches_summary.csv")

def export_modelling_approaches_to_csv(base_dir, output_path):
    data_to_export = []

    # On parcourt les dossiers de catégories (ex: covid_files)
    for category_dir in base_dir.iterdir():
        if category_dir.is_dir():
            
            # On parcourt les dossiers de modèles (ex: BIOMD...)
            for model_dir in category_dir.iterdir():
                if model_dir.is_dir():
                    
                    metadata_dir = model_dir / "metadata"
                    model_id = model_dir.name
                    disease_category = category_dir.name
                    
                    if metadata_dir.exists():
                        # Recherche du fichier JSON de métadonnées
                        json_files = list(metadata_dir.glob("*_web_metadata.json"))
                        
                        for json_file in json_files:
                            try:
                                with open(json_file, 'r', encoding='utf-8') as f:
                                    content = json.load(f)
                                    
                                    # Extraction des informations
                                    approach_data = content.get("modellingApproach", {})
                                    approach_name = approach_data.get("name", "Not specified")
                                    approach_uri = approach_data.get("resource", "N/A")
                                    
                                    # On ajoute une ligne à notre liste
                                    data_to_export.append({
                                        "disease_category": disease_category,
                                        "model_id": model_id,
                                        "modelling_approach": approach_name,
                                        "ontology_uri": approach_uri
                                    })
                                    
                            except Exception as e:
                                print(f"Erreur sur le fichier {json_file.name}: {e}")

    # Écriture du fichier CSV
    if data_to_export:
        keys = data_to_export[0].keys()
        with open(output_path, 'w', newline='', encoding='utf-8') as f_csv:
            writer = csv.DictWriter(f_csv, fieldnames=keys)
            writer.writeheader()
            writer.writerows(data_to_export)
        print(f"Succès : Fichier généré ici -> {output_path}")
    else:
        print("Aucun donnée trouvée à exporter.")

if __name__ == "__main__":
    if root_path.exists():
        export_modelling_approaches_to_csv(root_path, output_file)
    else:
        print(f"Le répertoire {root_path} est introuvable.")
```

```python
'''
publication date model
'''
import json
import csv
from pathlib import Path

# Configuration
root_path = Path("./BioModels_Data_cleaned_json_dico")  # Chemin vers vos données
output_file = Path("publication_dates_summary.csv")

def extract_publication_dates(base_dir, output_path):
    data_to_export = []

    # Parcours des dossiers de maladies
    for category_dir in base_dir.iterdir():
        if category_dir.is_dir():
            
            # Parcours des dossiers de modèles
            for model_dir in category_dir.iterdir():
                if model_dir.is_dir():
                    
                    metadata_dir = model_dir / "metadata"
                    model_id = model_dir.name
                    disease_category = category_dir.name
                    
                    if metadata_dir.exists():
                        # Recherche du fichier JSON contenant les métadonnées
                        json_files = list(metadata_dir.glob("*_web_metadata.json"))
                        
                        for json_file in json_files:
                            try:
                                with open(json_file, 'r', encoding='utf-8') as f:
                                    content = json.load(f)
                                    
                                    # Extraction de la section publication
                                    pub_data = content.get("publication", {})
                                    year = pub_data.get("year", "N/A")
                                    month = pub_data.get("month", "N/A")
                                    journal = pub_data.get("journal", "N/A")
                                    
                                    data_to_export.append({
                                        "disease_category": disease_category,
                                        "model_id": model_id,
                                        "publication_year": year,
                                        "publication_month": month,
                                        "journal": journal
                                    })
                                    
                            except Exception as e:
                                print(f"Error processing {json_file.name}: {e}")

    # Écriture du fichier CSV
    if data_to_export:
        keys = data_to_export[0].keys()
        with open(output_path, 'w', newline='', encoding='utf-8') as f_csv:
            writer = csv.DictWriter(f_csv, fieldnames=keys)
            writer.writeheader()
            writer.writerows(data_to_export)
        print(f"Success: File generated at {output_path}")
    else:
        print("No data found.")

if __name__ == "__main__":
    if root_path.exists():
        extract_publication_dates(root_path, output_file)
    else:
        print(f"Directory {root_path} not found.")
```

```python
'''suppression modeles avanr 2015'''

import json
import shutil
from pathlib import Path

# Configuration du chemin racine vers vos données
root_path = Path("./BioModels_Data_json_dico") 

def delete_old_models(base_dir, threshold_year=2015):
    deleted_count = 0
    kept_count = 0

    # On parcourt les catégories (covid_files, hiv_files, etc.)
    for category_dir in base_dir.iterdir():
        if not category_dir.is_dir():
            continue
            
        print(f"--- Processing category: {category_dir.name} ---")
        
        # On parcourt chaque dossier de modèle (BIOMD...)
        for model_dir in list(category_dir.iterdir()):
            if not model_dir.is_dir():
                continue
                
            metadata_file = model_dir / "metadata" / f"{model_dir.name}_web_metadata.json"
            should_delete = False
            
            if metadata_file.exists():
                try:
                    with open(metadata_file, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                        pub_year = data.get("publication", {}).get("year")
                        
                        # Si l'année est présente et inférieure au seuil
                        if pub_year and int(pub_year) < threshold_year:
                            should_delete = True
                        elif not pub_year:
                            print(f"No year found for {model_dir.name}, keeping by default.")
                            
                except Exception as e:
                    print(f"Error reading {model_dir.name}: {e}")
            else:
                print(f"Metadata missing for {model_dir.name}, skipping.")

            # Suppression du dossier
            if should_delete:
                print(f"Deleting {model_dir.name} (Published: {pub_year})")
                shutil.rmtree(model_dir)
                deleted_count += 1
            else:
                kept_count += 1

    print(f"\nCleanup finished!")
    print(f"Models deleted: {deleted_count}")
    print(f"Models kept (>= {threshold_year}): {kept_count}")

if __name__ == "__main__":
    if root_path.exists():
        # Attention : cette action est irréversible !
        confirm = input(f"Are you sure you want to delete models published before {2015}? (yes/no): ")
        if confirm.lower() == 'yes':
            delete_old_models(root_path)
        else:
            print("Operation cancelled.")
    else:
        print(f"Path {root_path} not found.")
```

```python
'''
code final
'''

import os
import shutil
import json
import time
import requests
import csv
from pathlib import Path
from bioservices import BioModels
import biomodels

# --- CONFIGURATION ---
RACINE_DATA = Path("./BioModels_Database_Final")
# Extensions redirected to metadata folder
EXTENSIONS_METADATA = {'.png', '.jpg', '.jpeg', '.pdf', '.txt', '.docx', '.doc', '.xlsx', '.xls', '.csv'}
# Extensions considered as core modeling files
EXTENSIONS_MODELE = {
    '.xml', '.sbml', '.omex', '.sedml', '.cps', '.m', '.ode', 
    '.py', '.f', '.java', '.vcml', '.zip', '.tsv', '.yaml', '.graphml'
}

DISEASES = {
    "dengue_files": ('dengue','DENV'),
    "chikungunya_files": ("chikungunya", "CHIKV"),
    "lyme_files": ('lyme', 'borrelia', 'borreliosis'),
    "mpox_files": ('mpox', 'monkeypox'),
    "west_nile_files": ('west nile', 'WNV'),
    "influenza_files": ("influenza", "influenza virus","avian influenza", "H5N1"),
    "tuberculosis_files": ("tuberculosis", "TB", "mycobacterium"),
    "hiv_files": ("HIV", "human immunodeficiency virus"),
    "covid_files": ("covid", "SARS-CoV-2")
}

# --- PIPELINE FUNCTIONS ---

def step_download_and_sort():
    """
    Step 1: Download models and perform initial sorting.
    Redirects metadata-like files to the 'metadata' folder and keeps modeling files in 'model'.
    Also deletes original 'metadata.*' files from the model folder if they are being replaced.
    """
    RACINE_DATA.mkdir(parents=True, exist_ok=True)
    s = BioModels()
    
    # Define browser-like headers to avoid being blocked by the server
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    for folder_name, queries in DISEASES.items():
        print(f"\n--- Category: {folder_name} ---")
        query_string = " OR ".join(f'"{q}"' for q in queries)
        results = s.search(query_string, numResults=100) 
        
        if not isinstance(results, dict):
            print(f"  ! Server error for {folder_name}. Waiting...")
            time.sleep(5)
            continue

        ids = [res['id'] for res in results.get('models', [])]
        time.sleep(1)

        for model_id in ids:
            try:
                # Setup directory structure for the model
                dossier_base = RACINE_DATA / folder_name / model_id
                d_metadata = dossier_base / "metadata"
                d_model = dossier_base / "model"
                
                d_metadata.mkdir(parents=True, exist_ok=True)
                d_model.mkdir(parents=True, exist_ok=True)

                # Files to preserve in the metadata folder (whitelist)
                files_to_preserve = [f"{model_id}_web_metadata.json", "files_list.json"]

                # 1. Download Web JSON Metadata (The generated reference file)
                print(f"  > Fetching Web JSON for {model_id}...")
                web_url = f"https://www.ebi.ac.uk/biomodels/{model_id}?format=json"
                try:
                    # Increased timeout to 20s for heavy models
                    web_res = requests.get(web_url, headers=headers, timeout=20)
                    if web_res.status_code == 200:
                        with open(d_metadata / files_to_preserve[0], "w", encoding="utf-8") as f_web:
                            json.dump(web_res.json(), f_web, indent=4)
                    else:
                        print(f"    ! Metadata download failed (Status: {web_res.status_code})")
                except Exception as e:
                    print(f"    ! Error fetching JSON: {e}")

                # 2. Retrieve the internal files list
                files_objects = biomodels.get_metadata(model_id)
                if not files_objects: 
                    continue
                
                with open(d_metadata / files_to_preserve[1], "w", encoding="utf-8") as f_meta:
                    json.dump(files_objects, f_meta, indent=4, default=str)

                # 3. Download and sort files based on extension and naming
                print(f"  > Sorting and downloading files for {model_id}...")
                for target_file_obj in files_objects:
                    try:
                        nom_reel = getattr(target_file_obj, 'name', str(target_file_obj))
                        if not nom_reel or nom_reel == "None": 
                            continue

                        # REMOVAL LOGIC: delete existing 'metadata.*' in 'model' folder to avoid duplicates
                        if nom_reel.lower().startswith("metadata."):
                            old_file_in_model = d_model / nom_reel
                            if old_file_in_model.exists():
                                old_file_in_model.unlink() 

                        # Extension analysis
                        ext = Path(nom_reel).suffix.lower() or "no_extension"

                        # Sorting logic: move to metadata if keywords match or extension is non-modeling
                        is_metadata_keyword = any(x in nom_reel.lower() for x in ["metadata", ".json", ".rdf", ".owl"])
                        is_non_modeling_ext = ext in EXTENSIONS_METADATA
                        
                        if is_metadata_keyword or is_non_modeling_ext:
                            dest_path = d_metadata / nom_reel
                            files_to_preserve.append(nom_reel) # Keep in whitelist
                        else:
                            dest_path = d_model / nom_reel

                        # Execute file download
                        result = biomodels.get_file(target_file_obj)
                        if isinstance(result, (str, Path)) and Path(result).exists():
                            shutil.copy(result, dest_path)
                        else:
                            with open(dest_path, "w", encoding="utf-8") as f:
                                f.write(str(result))
                    except Exception as e:
                        print(f"    ! Error processing file {nom_reel}: {e}")

                # 4. Final Cleanup: delete files in 'metadata' folder not in the whitelist
                for file_path in d_metadata.iterdir():
                    if file_path.name not in files_to_preserve:
                        file_path.unlink()

                print(f"  > {model_id} successfully processed.")
                time.sleep(1.0) # Pause to respect server limits

            except Exception as e:
                print(f"Critical error on model {model_id}: {e}")

def step_clean_and_stats():
    """Step 2: Remove models without modeling files and generate extension statistics"""
    stats_globales = []
    # Make sure we only iterate over existing directories
    if not RACINE_DATA.exists():
        print("Data directory not found.")
        return

    for dossier_maladie in [d for d in RACINE_DATA.iterdir() if d.is_dir()]:
        for dossier_model in list(dossier_maladie.iterdir()):
            if not dossier_model.is_dir(): continue
            
            chemin_model = dossier_model / "model"
            a_un_modele = False
            comptage = {}
            
            if chemin_model.exists():
                for f in chemin_model.iterdir():
                    if f.is_file():
                        ext = f.suffix.lower() if f.suffix else "no_ext"
                        comptage[ext] = comptage.get(ext, 0) + 1
                        if ext in EXTENSIONS_MODELE: a_un_modele = True

            # If no valid modeling file found, delete the entire model folder
            if not a_un_modele:
                print(f"[-] Deleting {dossier_model.name}: No valid modeling files found.")
                shutil.rmtree(dossier_model)
            else:
                row = {"category": dossier_maladie.name, "model_id": dossier_model.name}
                row.update(comptage)
                stats_globales.append(row)

    if stats_globales:
        # 1. Get all unique keys from all dictionaries
        all_keys = set().union(*(d.keys() for d in stats_globales))
        # 2. Remove the base columns using set subtraction
        extension_keys = all_keys - {'category', 'model_id'}
        # 3. Create the final ordered list
        cols = ['category', 'model_id'] + sorted(list(extension_keys))
        
        output_path = RACINE_DATA / "extension_stats_summary.csv"
        with open(output_path, "w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=cols)
            writer.writeheader()
            # Replace missing values with 0
            for row in stats_globales:
                writer.writerow({c: row.get(c, 0) for c in cols})
        
        print(f"Cleaning and statistics export completed: {output_path}")

def step_classify_by_approach():
    """Step 3: Reorganize folders by modeling approach and generate approach CSV"""
    data_export = []
    print("Reorganizing folders by modeling approach...")

    for category_dir in [d for d in RACINE_DATA.iterdir() if d.is_dir()]:
        for model_dir in list(category_dir.iterdir()):
            if not model_dir.is_dir(): continue
            
            json_path = model_dir / "metadata" / f"{model_dir.name}_web_metadata.json"
            approach_name = "Unknown_Approach"

            if json_path.exists():
                with open(json_path, 'r', encoding='utf-8') as f:
                    content = json.load(f)
                    approach_name = content.get("modellingApproach", {}).get("name", "Not_specified")
                    # Clean folder name for OS compatibility
                    folder_safe_name = approach_name.replace(" ", "_").replace("/", "-")
                    
                    data_export.append({
                        "disease_category": category_dir.name,
                        "model_id": model_dir.name,
                        "modelling_approach": approach_name
                    })

                # Move model directory to the approach-specific subfolder
                new_parent_dir = category_dir / folder_safe_name
                new_parent_dir.mkdir(exist_ok=True)
                try:
                    shutil.move(str(model_dir), str(new_parent_dir / model_dir.name))
                except Exception as e:
                    print(f"Could not move {model_dir.name}: {e}")

    if data_export:
        with open(RACINE_DATA / "modelling_approaches_summary.csv", "w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=data_export[0].keys())
            writer.writeheader()
            writer.writerows(data_export)
        print("Physical classification and CSV summary completed.")

def step_delete_pre_2015():
    """Step 4: Permanently delete models published before 2015"""
    confirm = input("Are you sure you want to delete models published before 2015? (yes/no): ")
    if confirm.lower() != 'yes': return

    # Recursive search for metadata files to check publication date
    for model_path in RACINE_DATA.glob("**/*_web_metadata.json"):
        # parent[0] = metadata, parent[1] = model ID folder
        model_dir = model_path.parents[1]
        try:
            with open(model_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                year = data.get("publication", {}).get("year")
                if year and int(year) < 2015:
                    print(f"Deleting {model_dir.name} (Published in: {year})")
                    shutil.rmtree(model_dir)
        except:
            continue

def step_separate_curation_status():
    """
    Step 5: Physically separate models into 'curated' and 'non_curated' subfolders
    within each disease category.
    """
    print("Separating models by curation status...")
    
    # We iterate through each disease category folder
    for category_dir in [d for d in RACINE_DATA.iterdir() if d.is_dir()]:
        # We search for metadata files deeply because they might already be in approach folders
        for json_path in list(RACINE_DATA.glob(f"{category_dir.name}/**/*_web_metadata.json")):
            # model_dir is two levels up from the JSON file (model_id/metadata/file.json)
            model_dir = json_path.parents[1]
            
            try:
                with open(json_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    # Get status, default to 'UNKNOWN' if missing
                    status = data.get("curationStatus", "UNKNOWN").lower()
                
                # Create the status subfolder (curated or non_curated)
                # We place it directly under the disease category
                target_parent_dir = category_dir / status
                target_parent_dir.mkdir(exist_ok=True)
                
                # Move the model folder to its new location
                # Example: covid_files/curated/BIOMD...
                new_location = target_parent_dir / model_dir.name
                
                if model_dir != new_location:
                    shutil.move(str(model_dir), str(new_location))
                    
            except Exception as e:
                print(f"  ! Could not process curation status for {model_dir.name}: {e}")

    print("Models have been sorted into 'curated' and 'non_curated' folders.")

# --- MAIN MENU ---

def main():
    while True:
        print("\n--- BIOMODELS ANALYSIS PIPELINE ---")
        print("1. Download & Sort (Metadata vs Model)")
        print("2. Clean Empty Models & Generate Extension Stats CSV")
        print("3. Classify Folders by Approach & Generate Approach CSV")
        print("4. Delete Models Published Before 2015")
        print("5. Separate by Curation Status (Curated vs Non-Curated)")
        print("6. Exit")
        
        choice = input("\nSelect an option (1-6): ")
        
        if choice == '1': step_download_and_sort()
        elif choice == '2': step_clean_and_stats() # Remember to use the fixed version here
        elif choice == '3': step_classify_by_approach()
        elif choice == '4': step_delete_pre_2015()
        elif choice == '5': step_separate_curation_status()
        elif choice == '6': break
        else: print("Invalid entry.")

if __name__ == "__main__":
    main()
```
